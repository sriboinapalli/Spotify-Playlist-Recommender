```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json("Frictionless.json")
```

```python
df.head()
df = df.set_index("id")
```

```python
df.head()
```

```python
df = df.drop(["analysis_url","track_href","uri","type"],1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.boxplot("acousticness")
```

```python
god = pd.read_json("I'm a god.json")
god = god.set_index("id")
god.head()
```

```python
god.describe()
```

```python
color = ["#BAA7B0","#FF9C6A"]
sns.set_style("white")
palette = sns.color_palette(color)
sns.set_palette(palette)
```

```python
fig = plt.figure(figsize = (15,15))

# Acousticness Distribution
ax1 = fig.add_subplot(3,3,1)
ax1.set_xlabel("Acousticness")
ax1.set_ylabel("Count")
ax1.set_title("Acousticness Distribution")
df['acousticness'].hist(alpha = .7, label = "Frictionless")
god['acousticness'].hist(alpha = .7, label = "I'm a God")
# Danceability Distribution
ax2 = fig.add_subplot(3,3,2)
ax2.set_xlabel("Danceability")
ax2.set_ylabel("Count")
ax2.set_title("Danceability Distribution")
df['danceability'].hist(alpha = .7)
god['danceability'].hist(alpha = .7, label = "I'm a God")
# Duration Distribution
ax3 = fig.add_subplot(3,3,3)
ax3.set_xlabel("Duration (ms)")
ax3.set_ylabel("Count")
ax3.set_title("Duration Distribution")
df['duration_ms'].hist(alpha = .7, label = "Frictionless")
god['duration_ms'].hist(alpha = .7, label = "I'm a God")
# Energy Distribution
ax4 = fig.add_subplot(3,3,4)
ax4.set_xlabel("Energy")
ax4.set_ylabel("Count")
ax4.set_title("Energy Distribution")
df['energy'].hist(alpha = .7, label = "Frictionless")
god['energy'].hist(alpha = .7, label = "I'm a God")
# Instrumentalness Distribution
ax5 = fig.add_subplot(3,3,5)
ax5.set_xlabel("Instrumentalness")
ax5.set_ylabel("Count")
ax5.set_title("Instrumentalness Distribution")
df['instrumentalness'].hist(alpha = .7, label = "Frictionless")
god['instrumentalness'].hist(alpha = .7, label = "I'm a God")
# Liveness Distribution
ax6 = fig.add_subplot(3,3,6)
ax6.set_xlabel("Liveness")
ax6.set_ylabel("Count")
ax6.set_title("Liveness Distribution")
df['liveness'].hist(alpha = .7, bins=15, label = "Frictionless")
god['liveness'].hist(alpha = .7, bins=15, label = "I'm a God")
# Loudness Distribution
ax7 = fig.add_subplot(3,3,7)
ax7.set_xlabel("Loudness")
ax7.set_ylabel("Count")
ax7.set_title("Loudness Distribution")
df['loudness'].hist(alpha = .7, bins = 15, label = "Frictionless")
god['loudness'].hist(alpha = .7, bins = 15, label = "I'm a God")
# Speechiness Distribution
ax8 = fig.add_subplot(3,3,8)
ax8.set_xlabel("Speechiness")
ax8.set_ylabel("Count")
ax8.set_title("Speechiness Distribution")
df['speechiness'].hist(alpha = .7, bins=15, label = "Frictionless")
god['speechiness'].hist(alpha = .7, bins = 15, label = "I'm a God")
# Valence Distribution
ax9 = fig.add_subplot(3,3,9)
ax9.set_xlabel("Valence")
ax9.set_ylabel("Count")
ax9.set_title("Valence Distribution")
df['valence'].hist(alpha = .7, bins=15, label = "Frictionless")
god['valence'].hist(alpha = .7, bins=15, label = "I'm a God")
```
